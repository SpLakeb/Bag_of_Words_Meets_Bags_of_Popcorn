```python
from bs4 import BeautifulSoup

example1 = BeautifulSoup(train['review'][0])

print(train['review'][0])
print(example1.text)
```

```python
words = [w for w in words if not w in stopwords.words("english")]
print(words)
```

```python
def review_to_words(raw_review):
    review_text = BeautifulSoup(raw_review).get_text()
    letters_only = re.sub('[^a-zA-Z]', " ", review_text)
    words = letters_only.lower().split()

    stop = stopwords.words('english')
    meaningful_words = [w for w in words if not w in stop]
    return (" ".join(meaningful_words))
```

```python
clear_review = review_to_words(train['review'][0])
print(clear_review)
```

```python
num_reviews = train['review'].size

clean_train_reviews = []
print("Cleaning and parsing the training set movie reviews...\n")
for i in range(0, num_reviews):
    if (i + 1) % 100 == 0:
        print("Review %d of %d\n" % (i + 1, num_reviews))
    clean_train_reviews.append(review_to_words(train['review'][i]))
```

```python
len(clean_train_reviews)
```

```python
print("create the bag of words...\n")

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer="word",
                             tokenizer=None,
                             preprocessor=None,
                             stop_words=None,
                             max_features=10000)

train_data_features = vectorizer.fit_transform(clean_train_reviews)

train_data_features = train_data_features.toarray()
```

```python
train_data_features.shape
```

```python
train_data_features[0]
```

```python
vocab = vectorizer.get_feature_names_out()
print(vocab.shape)
```

```python
import numpy as np

dist = np.sum(train_data_features, axis=0)
for tag, count in zip(vocab, dist):
    print(count, tag)
```

```python
print("Training the random forest...")
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=300)

forest = forest.fit(train_data_features, train["sentiment"])
```

```python
test = pd.read_csv(r"data/testData.tsv", header=0, delimiter="\t", quoting=3)
print(test.shape)
test
```

```python
num_reviews = len(test["review"])
clean_test_review = []

print("clean test review...\n")
for i in range(0, num_reviews):
    if (i + 1) % 100 == 0:
        print("review %d of %d\n" % (i + 1, num_reviews))
    clean_review = review_to_words(test["review"][i])
    clean_test_review.append(clean_review)

test_data_features = vectorizer.transform(clean_test_review)
test_data_features = test_data_features.toarray()
```

```python
result = forest.predict(test_data_features)
```

```python
output = pd.DataFrame(data={"id":test["id"], "sentiment":result})
output.to_csv("Bag_of_Words_model.csv", index=False, quoting=3)
```
